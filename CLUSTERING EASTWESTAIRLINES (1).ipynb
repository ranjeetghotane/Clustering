# **Clustering**
# Hierarchial

```python
# Normalization = (x-x(mean))/(x(max)-x(min))

def norm_func (n):
    g = (n-n.min())/(n.max()-n.min())
    return g
```

```python
airlinesf = norm_func(airlines)
```

```python
airlinesf.head(10)
```

```python
##Create the dendogram
plt.figure(figsize=(40,40))
dendrogram = sch.dendrogram(sch.linkage(airlinesf, method='complete'))
```

```python
hcc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
```

```python
hcc
```

```python
hcc.fit(airlinesf)
```

```python
hcc_y = hcc.fit_predict(airlinesf)
```

```python
hcc_y
```

```python
airlinesf['cluster ID'] = hcc_y
```

```python
airlinesf
```

```python
airlinesf.groupby('cluster ID').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(airlinesf['cluster ID'],airlinesf['Balance'], c=hcc_y)
```

# Kmeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
# Normalization

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# minmaxscalar
scaled_airlines = scaler.fit_transform(airlines.iloc[:,1:])
```

```python
scaled_airlines
```

```python
for i in range(1,4):
    print(i)
```

```python
kmeans = KMeans(n_clusters=4,random_state=0)
kmeans.fit(scaled_airlines)
inertia = []
inertia.append(kmeans.inertia_)
```

```python
inertia
```

```python
kmeans.inertia_
```

```python
inertia = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_airlines)
    inertia.append(kmeans.inertia_)
    if i >2 :
        break
```

```python
inertia
```

```python
inertia = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_airlines)
    inertia.append(kmeans.inertia_)

plt.plot(range(1,11),inertia)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# Building clustering algorithm
clusters_new = KMeans(5, random_state=42)
clusters_new.fit(scaled_airlines)
```

```python
clusters_new.labels_
```

```python
airlinesf['KMeans cluster ID'] = clusters_new.labels_
```

```python
airlinesf.head(10)
```

```python
#Clusterwise Grouping og mean of dataset

airlinesf.groupby('KMeans cluster ID').agg(['mean']).reset_index()
```

```python
(airlinesf[airlinesf['KMeans cluster ID']==0]).value_counts().sum()
```

```python
for i in range(0,5):
    n = airlinesf[airlinesf['KMeans cluster ID']==i].value_counts().sum()
    print('Cluster No','-', i, ';' , 'Count','-',n)
    print('_____________________________________________________________________________')
    print('\n', airlinesf[airlinesf['KMeans cluster ID']==i] )
    print('_____________________________________________________________________________')
```

# DBSCAN

```python
#DBSCAN LIBRARIES
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
array = airlinesf.values
array
```

```python
ss = StandardScaler()
x = ss.fit_transform(array)
```

```python
x
```

```python
dbscan = DBSCAN(eps=3, min_samples=3)
dbscan.fit_predict(x)
```

```python
dbscan.labels_
```

```python
cl = pd.DataFrame(dbscan.labels_,columns=['DBSCAN cluster'])
cl
```

```python
# airlinesf = pd.concat(['DBSCAN Clusters']) = cl 

airlinesf['DBSCAN Clusters'] = dbscan.labels_
```

```python
airlinesf
```

```python

for i in range(0,2):
    n = airlinesf[airlinesf['DBSCAN Clusters']==i].value_counts().sum()
    print('Cluster No','-', i, ';' , 'Count','-',n)
    print('_____________________________________________________________________________')
    print('\n', airlinesf[airlinesf['DBSCAN Clusters']==i] )
    print('_____________________________________________________________________________')
```
